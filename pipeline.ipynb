```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

```python
steps = [('standard_scaler',StandardScaler()),('LogisticRegression',LogisticRegression())]
```

```python
steps
```

```python
pipe = Pipeline(steps)
pipe
```

```python
from sklearn.datasets import make_classification
```

```python
x,y = make_classification(n_samples=123457)
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
model = pipe.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.impute import SimpleImputer
import numpy as np
from sklearn.preprocessing import OneHotEncoder
```

```python
numeric_transformer = Pipeline(steps = [('imputation', SimpleImputer(missing_values=np.nan, strategy=np.mean)), ('standard_scaler', StandardScaler())])
```

```python
numeric_transformer
```

```python
categoric = Pipeline(steps=[('imputation_constant', SimpleImputer(fill_value='missing', strategy='constant')), ('onhotencoder',OneHotEncoder())])
```

```python
categoric
```

```python
from sklearn.compose import ColumnTransformer
preprocessor = ColumnTransformer([
    ("categorical", categoric, ["gender", "City"]),
    ("numerical", numeric_transformer, ["age", "height"])
])
```

```python
preprocessor
```

```python
model = Pipeline(steps=[("preprocessor",preprocessor),("classifier",LogisticRegression())])
```
